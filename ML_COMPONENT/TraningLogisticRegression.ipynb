#Data can be generated by just running a random num generator with given range or made manually (not recommended)



```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
import numpy as np

#Loading dataset
# Expected columns:
#   - failed_attempts
#   - button_presses
#   - time_since_unlock
#   - >=1 indicates suspicious behavior
df = pd.read_csv("./Dataset/ML.csv")

#Just using 3 classes
X = df[['failed_attempts', 'button_presses', 'time_since_unlock']]

# Extract the multiclass labels.
y = df['class']

# Convert labels to a binary target for anomaly/suspicious behavior detection.
# Greater or equal to 1 means that the behavior is sus
y_binary = (y >= 1).astype(int)

#Load model as outlined in the documentation
model = LogisticRegression()

# Train model
model.fit(X, y_binary)

np.set_printoptions(precision=3)


# The learned coefficients weights from the formula.
print("Weights ->>> ", model.coef_)

#Learned bias term
print("B = Bias ->>>> ", model.intercept_)


model.score(X,y_binary)

test = [[1,90,6000]]
model.predict(test)
```
